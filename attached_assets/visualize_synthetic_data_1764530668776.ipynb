```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/smartgrid_synthetic.csv')
print(df.head())
print(df['attack_label'].value_counts())

plt.figure(figsize=(10,4))
sns.lineplot(x='time', y='Bus1_V', data=df, label='Bus 1 Voltage')
sns.lineplot(x='time', y='frequency', data=df, label='Frequency')
plt.title('Smart Grid Voltage and Frequency Over Time')
plt.legend()
plt.show()
```
